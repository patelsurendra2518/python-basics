```python

num = int(input("Enter a number: "))


if num % 2 == 0:
    print("The number is Even")
else:
    print("The number is Odd")
```

```python

age = int(input("Enter your age: "))
if age >= 0:
    if age < 13:
        print("Child")
    else:
        if age < 20:
            print("Teenager")
        else:
            if age < 60:
                print("Adult")
            else:
                print("Senior Citizen")
else:
    print("Invalid age")
```
